```python
def load_decoder_inputs(decoder_np_vecs='hindi_train_vecs.npy'):
    vectorized_title=np.load(decoder_np_vecs)
    decoder_input_data=vectorized_title[:,:-1]
    decoder_target_data=vectorized_title[:,1:]
    print(f'Shape of the decoder input:{decoder_input_data.shape}')
    print(f'Shape of the decoder target :{decoder_target_data.shape}')
    return decoder_input_data,decoder_target_data
```

```python
decoder_input_data,decoder_target_data=load_decoder_inputs('hindi_train_vecs.npy')
```

```python
def load_encoder_inputs(encoder_np_vecs='english_train_vecs.npy'):
    vectorized_body=np.load(encoder_np_vecs)
    encoder_input_data=vectorized_body
    doc_length=encoder_input_data.shape[1]
    print(f'Shape of the encoder input:{encoder_input_data.shape}')
    return encoder_input_data ,doc_length
```

```python
encoder_input_data,doc_length=load_encoder_inputs('english_train_vecs.npy')
```

```python
from keras.models import Model
from keras.layers import Input,LSTM,GRU,Dense,Embedding,Bidirectional,BatchNormalization
from keras import optimizers
```

#Model Architecture

```python
latent_dim=300
encoder_inputs=Input(shape=(doc_length,),name='Encoder-Input')
x=Embedding(number_of_input_token+1,latent_dim,name='Body-Word-Embedding',mask_zero=False)(encoder_inputs)
x=BatchNormalization(name='Encoder-Batchnorm-1')(x)
_,state_h=GRU(latent_dim,return_state=True,name='Encoder-Last-GRU')(x)
encoder_model=Model(inputs=encoder_inputs,outputs=state_h,name='Encoder-Model')
seq2seq_encoder_out=encoder_model(encoder_inputs)


decoder_input=Input(shape=(None,),name='Decoder-Input')
dec_emb=Embedding(number_of_output_token+1,latent_dim,name='Decoder-Word-Embedding',mask_zero=False)(decoder_input)
dec_bn=BatchNormalization(name='Decoder-Batchnorm-1')(dec_emb)

decoder_gru=GRU(latent_dim,return_state=True,return_sequences=True,name='Decoder-GRU')
decoder_gru_output,_=decoder_gru(dec_bn,initial_state=seq2seq_encoder_out)
x=BatchNormalization(name='Decoder-Batchnorm-2')(decoder_gru_output)

decoder_dense=Dense(number_of_output_token+1,activation='softmax',name='Final-Output-Dense')
decoder_outputs=decoder_dense(x)


seq2seq_Model=Model([encoder_inputs,decoder_input],decoder_outputs)
seq2seq_Model.compile(optimizer=optimizers.Nadam(lr=0.001),loss='sparse_categorical_crossentropy',)
```

```python
def extract_encoder_model(model):
  encoder_model=model.get_layer('Encoder-Model')
  return encoder_model

def extract_decoder_model(model):
     latent_dim=model.get_layer('Decoder-Word-Embedding').output_shape[-1]

     decoder_input=model.get_layer('Decoder-Input').input
     dec_emb=model.get_layer('Decoder-Word-Embedding')(decoder_input)
     dec_bn=model.get_layer('Decoder-Batchnorm-1')(dec_emb)



     gru_inference_state_input=Input(shape=(latent_dim,), name='hidden_states_input')

     gru_out,gru_state_out=model.get_layer('Decoder-GRU')([dec_bn,gru_inference_state_input])

     dec_bn2=model.get_layer('Decoder-Batchnorm-2')(gru_out)
     dense_out=model.get_layer('Final-Output-Dense')(dec_bn2)
     decoder_model=Model([decoder_input,gru_inference_state_input],[dense_out,gru_state_out])
     return decoder_model
```

# Model Inference Class

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np


seq2seq_model = load_model('/content/seq2seq.epoch23-val3.60886.hdf5')

# Extracting the encoder and decoder models
encoder_model = extract_encoder_model(seq2seq_model)
decoder_model = extract_decoder_model(seq2seq_model)
max_input_length=7
max_output_length=100
# Defining a function for translation using the decoder model
def translate(input_text):
    input_seq = Eng_token.texts_to_sequences([input_text])
    input_seq = pad_sequences(input_seq, maxlen=max_input_length, padding='post')

    encoder_out = encoder_model.predict(input_seq)
    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = Hindi_token.word_index['<start>']

    decoded_sentence = ''

    while True:
        output_tokens, h = decoder_model.predict([target_seq, encoder_out])

        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_token = Hindi_token.index_word[sampled_token_index]

        if sampled_token != '<end>':
            decoded_sentence += sampled_token + ' '

        target_seq = np.zeros((1, 1))
        target_seq[0, 0 ] = sampled_token_index

        encoder_out = h

        if sampled_token == '<end>' or len(decoded_sentence.split()) > max_output_length:
            break

    return decoded_sentence



# Example usage:
input_text = "Worship period "
translation = translate(input_text)
print(f"Input: {input_text}")
print(f"Translation: {translation}")
```
